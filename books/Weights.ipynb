```python
import numpy as np
import cvxpy as cvx

def cvx2np(x):
    return np.array([x[i].value for i in range(x.size[0])])


def minimize(objective, constraints=None):
    return cvx.Problem(cvx.Minimize(objective), constraints).solve()


def maximize(objective, constraints=None):
    return cvx.Problem(cvx.Maximize(objective), constraints).solve()


def std(vector):
    # we assume here that the vector is centered
    return cvx.norm(vector, 2)/np.sqrt(vector.size[0]-1)
```

```python
def elasticNet(matrix, w0, lamb1=0, lamb2=0):
    def cvx2np(x):
        return np.array([x[i].value for i in range(x.size[0])])

    def objective(matrix, lamb1=0, lamb2=0):
        return cvx.norm(matrix.values*w,2) + lamb1*cvx.norm(w,2) + lamb2*cvx.norm(w-w0,1)

    def minimize(objective, constraints=None):
        return cvx.Problem(cvx.Minimize(objective), constraints).solve()


    w = cvx.Variable(matrix.shape[1])
    
    constraints = [0 <= w, cvx.sum_entries(w) == 1]

    minimize(objective=objective(matrix, lamb1, lamb2), constraints=constraints)

    return 100*pd.Series(index=matrix.keys(), data=cvx2np(w)).sort_values()

elasticNet(matrix, w0, lamb1=0.0, lamb2=0.03)
```
